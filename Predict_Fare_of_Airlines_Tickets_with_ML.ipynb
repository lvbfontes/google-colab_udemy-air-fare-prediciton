**Checando tipos de dados e alterando os campos desejados para tipo data**

```python
train_data.dtypes
```

```python
def change_into_datetime(col):
  train_data[col] = pd.to_datetime(train_data[col])
```

```python
train_data.columns
```

```python
for i in ['Date_of_Journey','Dep_Time', 'Arrival_Time']:
  change_into_datetime(i)
```

```python
train_data.dtypes
```

**Extraindo dia e mes do campo 'Date_of_Journey' e removendo em seguida**

```python
train_data['journey_day'] = train_data['Date_of_Journey'].dt.day
train_data['journey_month'] = train_data['Date_of_Journey'].dt.month
```

```python
train_data.head()
```

```python
train_data.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
train_data.head()
```

```python
def extract_hour(df, col):
  df[col + '_hour'] = df[col].dt.hour

def extract_minute(df, col):
  df[col + '_minute'] = df[col].dt.minute

def drop_column(df, col):
  df.drop(col, axis=1, inplace=True)
```

```python
extract_hour(train_data, 'Dep_Time')
extract_minute(train_data, 'Dep_Time')
drop_column(train_data, 'Dep_Time')
```

```python
extract_hour(train_data, 'Arrival_Time')
extract_minute(train_data, 'Arrival_Time')
drop_column(train_data, 'Arrival_Time')
```

```python
train_data.head()
```

**Padronizando o formato da coluna 'Duration'**

```python
duration=list(train_data['Duration'])
duration
```

```python
#exemplo de split
x='2h 50m'
x.split(' ')
```

```python
len(x.split(' '))
```

```python
#criando loop para padronizar o formato com base no length de cada registro

for i in range(len(duration)):
  if len(duration[i].split(' '))==2:
    pass
  else:
    if 'h' in duration[i]:
      duration[i] = duration[i] + ' 0m'
    else:
      duration[i] = '0h ' + duration[i]

duration
```

```python
train_data['Duration']=duration
```
